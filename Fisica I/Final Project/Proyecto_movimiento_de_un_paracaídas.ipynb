# La velocidad terminal en la trayectoria de un paracaídas

Con base en los datos obtenidos en Tracker, se analiza el periodo de tiempo en el que la velocidad durante la caída se mantiene constante y de esta forma se promedia ese intervalo para obtener un valor de velocidad terminal. Teniendo el valor de la velocidad terminal para cada área, queda hallar el valor de la constante k y graficar los resultados para determinar la relación de proporción, si la hay


```python
#Datos completos obtenidos con tracker
import matplotlib.pyplot as plt
import math
import numpy as np

area= [0.0225,0.04,0.0625,0.09,0.1225,0.16,0.2025,0.25,0.3025,0.36]
vel_t = [4.770,3.627,2.916,2.709,1.933,1.789,1.448,1.204,1.072,0.987]
k = [9.81/v**2 for v in vel_t]

a = [const*0.013 for const in k]
fd = []
for i in range(len(vel_t)):
  fd.append(a[i]*(vel_t[i]**2))


plt.rcParams["figure.figsize"] = [10,10]

plt.subplot(2,2,1)
plt.plot(area, vel_t)
plt.xlabel('área, (m2)')
plt.ylabel('velocidad terminal (m/s)')
plt.grid()
plt.title('Velocidad terminal en función del área')

plt.subplot(2,2,2)
plt.plot(area, k)
plt.xlabel('área, m2')
plt.ylabel('constante k, 1/m')
plt.grid()
plt.title('constante k en función del área')

#plt.show()
```

## Relación de la constante $k$ y el área del paracaídas

Es posible evidenciar una clara relación de proporcionalidad directa. Es decir, a medida que el área aumente, la constante $k$ también aumenta. Sin embargo, esto es posible entenderlo a mayor detalle obteniendo la relación entre área y velocidad terminal, puesto que esta última es la que determina directamente la constante $k$

## Relación de la velocidad terminal y el área del paracaídas

En este caso es posible evidenciar una relación inversa. Es decir, a medida que el área del paracaídas aumenta, la velocidad terminal que alcanza es menor. Finalmente, es posible determinar que mientras la velocidad terminal sea menor, la constante $k$ será mayor.

# Análisis y comparación de las gráficas de velocidad en función del tiempo para cada área del paracaídas

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def velocity(dt,vt,nstep):
  v1 = []
  v2 = []
  t = []
  t_0 = 0
  v0 = 0
  t.append(t_0)
  v1.append(v0)
  v2.append(v0)
  g = 9.81
  k= 9.81/vt**2

  for i in range(0,nstep-1):
    t.append(t[i]+dt)
    vi1 = (math.sqrt(g/k))*math.tanh(t[i]*math.sqrt(g*k))
    v1.append(vi1)
    vi2 = vt*math.tanh((g*t[i])/vt)
    v2.append(vi2)

  return(t,v1,v2)
```

# Desarrollo computacional - numérico

Si consideramos el paracaídas como una partícula con una masa determinada $m$, es posible establecer que las fuerzas que actúan sobre él son el peso $F_g$ y la fuerza de fricción del aire $F_d$.

Por otro lado, se tiene que $$F_d=mg\tag{1}$$ donde $g$ es la constante gravitacional.

Para determinar $F_d$ se parte de que esta fuerza es proporcional a la velocidad de la partícula al cuadrado:
$$F_d \propto v^2$$
Por lo tanto:
$$F_d = Av*2\tag{3}$$

Donde $A$ es una constante de proporcionalidad.

Reemplazando $(1)$y $(2)$ para escribir la segunda ley de Newton, se obtiene:
$$F_g-F_d=ma$$
$$mg-Av^2=ma\tag{4}$$

Tomando la definición de la aceleración como primera derivada de la velocidad, podemos despejar y obtener:
$$\frac{dv}{dt}=g-kv^2$$
Donde $K=\frac{A}{m}$

Si tomamos la definición de la derivada, podemos desarrollar

$$\frac{v(t+\Delta t)-v(t)}{\Delta t}= g-kv^2$$
$$v(t+\Delta t)-v(t)= (g-kv^2)\Delta t$$
$$v_{i+1}= v_i+(g-kv^2)\Delta t \tag{5}$$
$$x_{i+1}=v_i\Delta t + x_i\tag{6}$$
​
Con $(5)$ y $(6)$ se tiene un proceso iterativo para hallar la velocidad y la posición, respectivamente.

Con esta solución numérica es posible determinar la velocidad y la posición del paracaídas a lo largo de su trayectoria.

Sin embargo, esto ya fue posible realizarlo con el desarrollo analítico. Ahora, será más relevante llevar a cabo una propuesta de un modelo predictivo, con base en los datos y tendencias obtenidas experimentalmente.

## Modelo predictivo

Sabemos que la constante $K$ tiene una relación con el área del paracaídas, de modo que cuando esta aumenta, la constante también lo hace. Pero, ¿de qué forma se da esta relación?

Para determinar esto, es posible realizar un ajuste polinomial a la gráfica que obtuvimos con los datos experimentales de $k$.

```python
#haciendo un ajuste polinomial de los datos de k
coef = np.polyfit(area,k,deg=2)

fit = [coef[0]*a**2+ coef[1]*a + coef[2] for a in area]
#fit = [coef[0]*a+ coef[1] for a in area]
plt.rcParams["figure.figsize"] = [5,5]
plt.plot(area, k)
plt.plot(area, fit)
plt.xlabel('área, m2')
plt.ylabel('constante k, 1/m')
plt.grid()
plt.title('constante k en función del área')
plt.show()
```

Es posible evidenciar que un polinomio de segundo grado permite una mejor aproximación a la curva real, por lo que se obtienen sus coeficientes. En este caso, se obtienen 3 coeficientes, por lo que el polinomio se expresa:
$$a(area^2)+b(area)+c$$
Donde $a$, $b$ y $c$ son los coeficientes obtenidos con la función, en el orden del array.

Teniendo esto, podemos entonces crear una función en la que, ingresando el área del paracaídas, sea posible determinar su velocidad terminal, además de su distancia recorrida.

```python
#@title Velocidad y posición del paracaídas en función de su área { run: "auto" }
area_m2 = 0.09 #@param {type:"number"}
def traj(a, coef):
  vel_l=[]
  pos_l = []
  t = [0]
  dt = 0.0001
  k= coef[0]*a**2+ coef[1]*a + coef[2]
  g= 9.81
  vel0=0
  pos0=0
  vel_l.append(vel0)
  pos_l.append(pos0)
  i=0
  while pos_l[i]<4.5:
  #for i in range (nstep-1):
    pos_l.append((vel_l[i]*dt)+ pos_l[i])
    vel_l.append(vel_l[i]+((g-k*((vel_l[i])**2))*dt))
    t.append(t[i]+dt)
    i=i+1
  return t, pos_l, vel_l
#import matplotlib.pyplot as plt

time, pos1, vel1 = traj(area_m2, coef)

plt.rcParams["figure.figsize"] = [12,12]
plt.subplot(2,2,1)
plt.plot(time, pos1)
plt.xlabel("Tiempo(s)")
plt.ylabel("posición (m)")
plt.title('posición vs tiempo')
plt.grid()

plt.subplot(2,2,2)
plt.plot(time, vel1)
plt.xlabel("Tiempo(s)")
plt.ylabel("velocidad (m/s)")
plt.title('velocidad vs tiempo')
plt.grid()
```
